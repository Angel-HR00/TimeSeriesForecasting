```python
# Your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf
from pmdarima import auto_arima


data = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/alternative-time-series-project/main/sales.csv")
data.head(6)
```

```python
data["date"] = pd.to_datetime(data.date)
data = data.set_index("date")

data_cl = data["sales"]
data_cl.head(6)
```

```python
fig, axis = plt.subplots(figsize = (10,7))

sns.lineplot(data = data_cl)
plt.show()
```

```python
def test(timeseries):
    print("Dikey-fuller test results:")
    df_test = adfuller(timeseries, autolag = "AIC")
    df_output = pd.Series(df_test[0:4], index = ["Test Statistic", "p-value", "#Lags Used", "Number of Observations Used"])
    for i,j in df_test[4].items():
        df_output["Critical value (%s)" %i] = j
    return df_output

test(data_cl)
```

```python
decomposition = seasonal_decompose(data_cl)
decomposition
```

```python
residual = decomposition.resid

fig, axis = plt.subplots(figsize = (10, 5))

sns.lineplot(data = data_cl)
sns.lineplot(data = residual)

plt.tight_layout()
plt.show()
```

```python
plot_acf(data_cl)

plt.tight_layout()

plt.show()
```

```python
model = auto_arima(data_cl, seasonal = False, trace = True, m = 7)
```

```python
forecast = model.predict(60)
forecast
```

```python
fig, axis = plt.subplots(figsize = (10, 5))

sns.lineplot(data = data_cl)
sns.lineplot(data = forecast, c = "green")

plt.tight_layout()

plt.show()
```
